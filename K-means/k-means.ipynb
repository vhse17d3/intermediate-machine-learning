Thuật toán k-means rất nhạy cảm với tỷ lệ. Điều này có nghĩa là chúng tôi cần phải suy nghĩ kỹ về cách thức và liệu chúng tôi có thay đổi quy mô các tính năng của mình hay không vì chúng tôi có thể nhận được các kết quả rất khác nhau tùy thuộc vào lựa chọn của mình. Theo nguyên tắc chung, nếu các tính năng đã có thể so sánh trực tiếp (chẳng hạn như kết quả kiểm tra ở các thời điểm khác nhau), thì bạn sẽ không muốn thay đổi tỷ lệ. Mặt khác, các tính năng không có thang đo tương đương (như chiều cao và cân nặng) thường sẽ được hưởng lợi từ việc thay đổi tỷ lệ. Đôi khi, sự lựa chọn sẽ không rõ ràng. Trong trường hợp đó, bạn nên cố gắng sử dụng ý thức chung, hãy nhớ rằng các tính năng có giá trị lớn hơn sẽ được đánh giá cao hơn.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.model_selection import cross_val_score
from xgboost import XGBRegressor

# Set Matplotlib defaults
plt.style.use("seaborn-whitegrid")
plt.rc("figure", autolayout=True)
plt.rc(
    "axes",
    labelweight="bold",
    labelsize="large",
    titleweight="bold",
    titlesize=14,
    titlepad=10,
)


def score_dataset(X, y, model=XGBRegressor()):
    # Label encoding for categoricals
    for colname in X.select_dtypes(["category", "object"]):
        X[colname], _ = X[colname].factorize()
    # Metric for Housing competition is RMSLE (Root Mean Squared Log Error)
    score = cross_val_score(
        model, X, y, cv=5, scoring="neg_mean_squared_log_error",
    )
    score = -1 * score.mean()
    score = np.sqrt(score)
    return score
df = pd.read_csv("./data/ames.csv")
```

# Create a Feature of Cluster Labels
Tạo cụm k-mean với các tham số sau:
- features: LotArea, TotalBsmtSF, FirstFlrSF, SecondFlrSF,GrLivArea
- number of clusters: 10
- iterations: 10

```python
X = df.copy()
y = X.pop("SalePrice")

features = [
    "LotArea",
    "TotalBsmtSF",
    "FirstFlrSF",
    "SecondFlrSF",
    "GrLivArea",
]

# Standardize
X_scaled = X.loc[:, features]
X_scaled = (X_scaled - X_scaled.mean(axis=0)) / X_scaled.std(axis=0)
"""
X.loc[:, features]: Lựa chọn các cột được chỉ định trong danh sách features.
(X_scaled - X_scaled.mean(axis=0)) / X_scaled.std(axis=0): 
Chuẩn hóa dữ liệu bằng cách trừ đi giá trị trung bình và chia cho độ lệch chuẩn. 
Điều này giúp đưa dữ liệu về cùng một phạm vi và giữ cho mỗi đặc trưng có trọng số tương đương trong mô hình.
"""
kmeans = KMeans(n_clusters=10, n_init=10, random_state=0)
X["Cluster"] = kmeans.fit_predict(X_scaled)
print(X["Cluster"])
```

```python
Xy = X.copy()
Xy["Cluster"] = Xy.Cluster.astype("category")
Xy["SalePrice"] = y
"""
Xy = X.copy(): Tạo một bản sao của DataFrame X để không ảnh hưởng đến DataFrame gốc.
Xy["Cluster"] = Xy.Cluster.astype("category"): Chuyển đổi cột "Cluster" thành kiểu dữ liệu category, phù hợp cho việc sử dụng nhãn của cụm trong biểu đồ.
Xy["SalePrice"] = y: Thêm cột "SalePrice" vào DataFrame Xy chứa giá bán (y), là biến phụ thuộc.
"""
sns.relplot(
    x="value", y="SalePrice", hue="Cluster", col="variable",
    height=4, aspect=1, facet_kws={'sharex': False}, col_wrap=3,
    data=Xy.melt(
        value_vars=features, id_vars=["SalePrice", "Cluster"],
    ),
);
```

```python
score_dataset(X, y) # tính toán mean dựa trên phân loại thuộc cụm 
```

# Cluster-Distance Features
Bây giờ hãy thêm các tính năng khoảng cách cụm vào tập dữ liệu của bạn. Bạn có thể nhận được các đặc điểm khoảng cách này bằng cách sử dụng phương thức fit_transform của kmeans thay vì fit_predict.

```python
kmeans = KMeans(n_clusters=10, n_init=10, random_state=0)

# YOUR CODE HERE: Create the cluster-distance features using `fit_transform`
X_cd = kmeans.fit_transform(X_scaled)

# Label features and join to dataset
X_cd = pd.DataFrame(X_cd, columns=[f"Centroid_{i}" for i in range(X_cd.shape[1])])
X = X.join(X_cd)
```

```python
score_dataset(X, y) # tính toán mean dựa trên bảng khoảng cách các tâm đã được phân loại
```
